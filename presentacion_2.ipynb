```python
import pandas as pd
import numpy as np
```

```python
atrasos = pd.read_csv('../dex/clean/modelo.csv', delimiter=';', low_memory=False)
atrasos
```

```python
atrasos['deuda_default'] = (atrasos['atr_61_90_dias']+
                             atrasos['atr_91_120_dias']+
                             atrasos['atr_121_150_dias']+
                             atrasos['atr_151_180_dias']+
                             atrasos['atr_181_360_dias']+
                             atrasos['atr_360_dias_a_mas'])
```

```python
def generando_cartera_atrasada(row, columnas):
    global valor_sumado
    valor_sumado = 0
    for i in columnas:
        if pd.notna(row[i]):
            valor_sumado = valor_sumado + row[i]
    return valor_sumado

columnas = ['atr_8_15_dias', 'atr_16_30_dias', 'atr_31_45_dias', 'atr_46_60_dias', 'atr_61_90_dias', 'atr_91_120_dias', 'atr_121_150_dias', 'atr_151_180_dias', 'atr_181_360_dias', 'atr_360_dias_a_mas']
atrasos = atrasos.assign(cartera_atrasada_mayor_7_dias=atrasos.apply(generando_cartera_atrasada, axis=1, columnas=columnas))
```

```python
data = atrasos[atrasos['var_mes']>202002]
data
```

```python
clientes = pd.read_csv('../dex/clientes.csv',delimiter=';')
clientes
```

```python
df = data.merge(clientes, left_on='dex_cliente_id', right_on='id', how='left')
df = df.drop(columns=['id'])
df
```

```python
dex = pd.read_csv('../dex/dex_info.csv', delimiter=';')
print(dex.info())
dex
```

```python
df = df.merge(dex, left_on='dex_cliente_id', right_on='customer_id', how='left')
df = df.drop_duplicates(subset=['var_mes','dex_cliente_id'])
df
```

```python
df = df[df['type'].isin(
    ['BODEGAS','PUESTO DE MERCADO','GASTRONOMÍA','PANIFICACIÓN','MAYORISTAS']
)]

df
```

```python
df['tipo'] = np.where(df['type'].isin(['GASTRONOMÍA','PANIFICACIÓN']),'B2B','CMP')
df['tipo'].value_counts()
```

```python
#CLIENTES POR MES
n_clientes = pd.pivot_table(
    df,
    index='var_mes',
    aggfunc={'dex_cliente_id':pd.Series.nunique}
)

n_clientes
```
