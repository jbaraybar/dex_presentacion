```python
df['tipo_dex'] = np.where((df['type'] == 'GASTRONOMÍA') | (df['type'] == 'PANIFICACIÓN'),'B2B','CMP')
#df['flag_default'] = np.where(df['deuda_default'] > 0,1,0)

df['year'] = df['var_mes'].astype(str).str.slice(stop=4)
```

```python
var_mes = list(df['var_mes'].unique())
bocas_salida = list(df['type'].unique())
```

```python
cant_default = []
deuda = []
boca_salida = []
mes = []

j = 5

for i in range(0, len(var_mes) - 1):
    customer_ant = df[
        (df.tramo >= j) & (df.var_mes == var_mes[i])
    ].dex_cliente_id.unique()
    for boca in bocas_salida:
        a = df[
            (df.tramo >= j)
            & (df.var_mes == var_mes[i + 1])
            & (~df.dex_cliente_id.isin(customer_ant)
            & (df.type == boca))
        ].shape[0]
        b = df[
            (df.tramo >= j)
            & (df.var_mes == var_mes[i + 1])
            & (~df.dex_cliente_id.isin(customer_ant)
            & (df.type == boca))
        ].deuda_default.sum()
        c = boca

        mes.append(var_mes[i + 1])
        cant_default.append(a)
        deuda.append(b)
        boca_salida.append(c)
```

```python
data = pd.DataFrame()
data["var_mes"] = mes
data["cant_default"] = cant_default
data["deuda"] = deuda
data["boca_de_salida"] = boca_salida
data['year'] = data['var_mes'].astype(str).str.slice(stop=4)

data['tipo_dex'] = np.where(data['boca_de_salida'].isin(['PANIFICACIÓN','GASTRONOMÍA']),'B2B','CMP')

data
```

```python
data.to_csv('../defaulters.csv')
```

```python
df_b2b = data[data['tipo_dex'] == 'B2B']
df_cmp = data[data['tipo_dex'] == 'CMP']
```

```python
data_sum = pd.DataFrame()
meses = []
default = []
deuda = []
tipos = []

for mes in data['var_mes'].unique():
    for tipo in data['tipo_dex'].unique():
        meses.append(mes)
        default.append(data['cant_default'][(data['var_mes'] == mes)&(data['tipo_dex'] == tipo)].sum())
        deuda.append(data['deuda'][(data['var_mes'] == mes)&(data['tipo_dex'] == tipo)].sum())
        tipos.append(tipo)

data_sum['var_mes'] = meses
data_sum['default'] = default
data_sum['deuda'] = deuda
data_sum['tipo'] = tipos
data_sum['year'] = data_sum['var_mes'].astype(str).str.slice(stop=4)

data_sum
```

```python
#Default Anual DEX
temp3 = pd.pivot_table(
    data_sum,
    index=['year'],
    columns=['tipo'],
    aggfunc={"deuda": "sum"},
)
temp3 = temp3.reset_index()
temp3
```
