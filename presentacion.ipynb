```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.float_format", lambda x: "%.2f" % x)
pd.set_option('mode.chained_assignment', None)
```

```python
atrasos = pd.read_csv('../dex/atrasos.csv', delimiter=';', low_memory=False)
atrasos.info()
```

```python
atrasos = atrasos[atrasos['payment_condition_id'] != '#']
```

```python
atrasos = atrasos.drop_duplicates(subset=['customer_dex_id','amount'], keep='first')
atrasos
```

```python
atrasos = atrasos.groupby(['month_id','customer_dex_id','customer_dex'])[['amount','on_time','delay_001_007_days',
                                                                          'delay_008_015_days','delay_016_030_days',
                                                                          'delay_031_060_days','delay_061_090_days',
                                                                          'delay_091_120_days','delay_121_150_days',
                                                                          'delay_151_180_days','delay_181_360_days',
                                                                          'delay_360_to_more']].sum()

atrasos = atrasos.reset_index()

atrasos
```

```python
clientes = pd.read_csv('../dex/clientes.csv', delimiter=';', low_memory=False)
clientes.info()
```

```python
df = atrasos.merge(clientes, left_on='customer_dex_id', right_on='id', how='left')
df
```

```python
df.columns
```

```python
df = df.drop(['id','name'],axis=1)
df = df[['month_id', 'customer_dex_id', 'customer_dex', 'type', 'channel', 'amount', 'on_time',
       'delay_001_007_days', 'delay_008_015_days', 'delay_016_030_days',
       'delay_031_060_days', 'delay_061_090_days', 'delay_091_120_days',
       'delay_121_150_days', 'delay_151_180_days', 'delay_181_360_days',
       'delay_360_to_more']]
df
```

```python
df[df['type'].isnull() | df['channel'].isnull()]
```

```python
df = df[df['type'].notna()]
```

```python
df['type'].value_counts()
```

```python
df = df[df['type'].isin(['BODEGAS','PUESTO DE MERCADO','GASTRONOMÍA','PANIFICACIÓN','MAYORISTAS'])]
```

```python
df['channel'].value_counts()
```

```python
df = df[df['month_id'].between(201801,202012)]

df['month_id'].unique()
```

```python
df.head()
```

```python
df[['customer_dex_id']] = df['customer_dex_id'].astype('int64')
df.iloc[:,5:] = df.iloc[:,5:].fillna(0)
```

```python
df.info()
```

```python
df['deuda_default'] = (df['delay_061_090_days']+
                       df['delay_091_120_days']+df['delay_121_150_days']+
                       df['delay_151_180_days']+df['delay_181_360_days']+
                       df['delay_360_to_more'])
```

```python
df
```

```python
df[df['deuda_default'] > 0]
```

```python
df = df[df['amount'] >=0]
df
```

```python
df['tipo_dex'] = np.where((df['type'] == 'GASTRONOMÍA') | (df['type'] == 'PANIFICACIÓN'),'B2B','CMP')
df['flag_default'] = np.where(df['deuda_default'] > 0,1,0)

df['year'] = df['month_id'].astype(str).str.slice(stop=4)
```

```python
df_b2b = df[df['tipo_dex'] == 'B2B']
df_cmp = df[df['tipo_dex'] == 'CMP']
```

```python
temp = pd.pivot_table(
    df,
    index=["month_id"],
    aggfunc={"amount": "sum",
             "deuda_default": "sum",
             "on_time": "sum"},
)
temp = temp.reset_index()
temp
```

```python
# clientes únicos por año
temp2 = pd.pivot_table(
    df[df['deuda_default'] > 0],
    index=["month_id"],
    aggfunc={"customer_dex_id": pd.Series.nunique},  # count distinct
)
temp2 = temp2.reset_index()
temp2
```
